```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
winedata = pd.read_csv('wine-clustering.csv')
```

```python
winedata.info()
```

```python
winedata.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
stddata = scaler.fit_transform(winedata)
```

```python
data = pd.DataFrame(stddata,columns= winedata.columns)
```

```python
data.info()
```

```python
data.describe()
```

```python
data=data.round(2)
```

```python
data.describe()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(data.cov(),linewidth=2,fmt='0.02f',annot=True)
```

```python
data.columns
```

```python
data = data.drop(['Hue','Nonflavanoid_Phenols','Proline','Flavanoids','OD280'],axis=1)
```

```python
data.info()
```

```python
sns.pairplot(data)
```

```python
from sklearn.cluster import KMeans
```

```python
cls = KMeans(n_clusters=5,n_init = 50,random_state=42)
```

```python
km = cls.fit(data)
```

```python
km.labels_
```

```python
km.inertia_
```

```python
clusters = []
for i in range(2,20):
    cls = KMeans(n_clusters=i,n_init=100,random_state=1334)
    km = cls.fit(data)
    clusters.append(km.inertia_)
clusters_series = pd.Series(clusters,index=range(2,20))
clusters_series
```

```python
sns.lineplot(y=clusters_series,x=clusters_series.index)
sns.scatterplot(y=clusters_series,x=clusters_series.index)
plt.xlabel('Value of K')
plt.ylabel('Inertia')
plt.show()
```

```python
from sklearn.metrics import silhouette_score as shscore
```

```python
clusters = []
for i in range(2,20):
    cls = KMeans(n_clusters=i,n_init=100,random_state=1334)
    km = cls.fit(data)
    clusters.append(shscore(data,km.labels_))
clusters_series = pd.Series(clusters,index=range(2,20))
clusters_series
```

```python
sns.lineplot(y=clusters_series,x=clusters_series.index)
sns.scatterplot(y=clusters_series,x=clusters_series.index)
plt.xlabel('Value of K')
plt.ylabel('Silhouette Score')
plt.show()
```
